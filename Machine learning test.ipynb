```python
import matplotlib.pyplot as plt
import seaborn as sns


graph = 1


plt.figure(figsize=(20, 15), facecolor='red')


for column in data.columns:
    if graph <= 9:  
       
        ax = plt.subplot(3, 3, graph)
        
        
        sns.histplot(data[column], kde=True, ax=ax) 
        
        
        ax.set_xlabel(column)
        ax.set_title(f'Distribution of {column}')
        
        graph += 1


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure
plt.figure(figsize=(20, 15), facecolor='yellow')

# Loop through the first 9 columns
for i, column in enumerate(data.columns[:9], 1):  # Only take the first 9 columns
    plt.subplot(3, 3, i)  # Create a 3x3 grid of subplots
    sns.histplot(data[column], kde=True)  # Plot the distribution
    plt.xlabel(column, fontsize=15)  # Label for each subplot

plt.tight_layout()  # Adjust spacing
plt.show()
```
